# Gaussian Distribution Fitting:

The equation for a normal (gaussian) distribution is:
$$f(x)=\frac{1}{\sqrt{2\pi \sigma^{2}}}exp(-\frac{(x-\mu)^{2}}{2\sigma ^{2}})$$


Where $\sigma$ is the standard deviation, found as:

$$\sigma = \sqrt{\frac{1}{N}\sum_{i=1}^{N}(x_{i}-\mu)^{2}}$$

and $\mu$ is the mean/average value, found as:

$$\mu = \frac{1}{N}\sum_{i=1}^{N}x_{i}$$

The value $\frac{1}{\sqrt{2\pi \sigma^{2}}}$ is the normalization factor, used to ensure the area under the curve is equal to 1.
This formula can be simplified to:
$$f(x)=A \cdot exp(-\frac{(x-\mu)^{2}}{2\sigma ^{2}})$$
Where A is the amplitude of the data.

To find the equation of the peaks, we should find the average of the plateau before the compton edge to act as the baseline, and use that value to find the amplitude of the peaks.

```python
# I'm not 100% sure about this code - I went back and forth with chatGPT on a decent amount of it
# Mainly unsure if the baseline should be used this way
# And also unsure why mu doesn't work as the mean value here


def gaussian(x, mu, sigma, amplitude):
    return amplitude * np.exp(-((x - mu)**2) / (2 * sigma**2))

def gaussian_fit(xdata, ydata, baseline_data):
    # Calculate average baseline
    baseline_av = np.mean(baseline_data)#np.min(baseline_data)
    print(f"Data is y-shifted by {baseline_av:.2f} counts")
    
    # Remove baseline from data
    ydata_corrected = ydata - baseline_av
    amplitude = np.max(ydata_corrected)
    
    # Estimate mu
    mu = xdata[np.argmax(ydata_corrected)]#np.mean(ydata_corrected)
    
    # Estimate sigma
    xi = 0
    for i in range(0, len(ydata)):
        xi_minus_mu = xdata[i] - mu
        xi += xi_minus_mu**2
        
    sigma = np.sqrt(xi/len(ydata_corrected))
    
    xaxis = []
    yaxis = []
    for i in range(0, len(xdata)):
        x = xdata[i]
        y_gauss = gaussian(x, mu, sigma, amplitude)
        yaxis.append(y_gauss)
        xaxis.append(xdata[i])
        
    return mu, sigma, amplitude, xaxis, yaxis
```

```python
x_data = (df_peak_1.iloc[:, 0].values)#/1000 # converting from KeV to MeV
y_data = df_peak_1.iloc[:, 1].values
#baseline_data = compton_plateau.iloc[:,1]
#baseline_data = df_plateau.iloc[:,1]
baseline_data = compton_cliff.iloc[:,1]
    
mu, sigma, amplitude, xaxis, yaxis = gaussian_fit(x_data, y_data, baseline_data)
print(f"mu: {mu:.2f}, sigma: {sigma:.2f}, amplitude: {amplitude:.2f} counts")

plt.scatter(x_data, y_data, color="blue", label="Data")
plt.plot(xaxis, yaxis, color="red", label="Fit")
plt.show()

print(f"The expected value of mu -the peak wavelength- is the known peak of 1.175 MeV. The calculated mu is {mu/1000:.2f} MeV")
print(f"This gives an x-axis shift by {mu-1175:.2f} KeV or {(mu-1175)/1000:.2f} MeV")
```

```python
# Confirming with peak 2
x_data = (df_peak_2.iloc[:, 0].values)#/1000 # converting from KeV to MeV
y_data = df_peak_2.iloc[:, 1].values
#baseline_data = compton_plateau.iloc[:,1]
#baseline_data = df_plateau.iloc[:,1]
baseline_data = compton_cliff.iloc[:,1]
    
mu, sigma, amplitude, xaxis, yaxis = gaussian_fit(x_data, y_data, baseline_data)
print(f"mu: {mu:.2f}, sigma: {sigma:.2f}, amplitude: {amplitude:.2f} counts")

plt.scatter(x_data, y_data, color="blue", label="Data")
plt.plot(xaxis, yaxis, color="red", label="Fit")
plt.show()

print(f"The expected value of mu -the peak wavelength- is the known peak of 1.333 MeV. The calculated mu is {mu/1000:.2f} MeV")
print(f"This gives an x-axis shift by {mu-1333:.2f} KeV or {(mu-1333)/1000:.2f} MeV")
```
