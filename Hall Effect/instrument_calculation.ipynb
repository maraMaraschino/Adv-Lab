# Calculating Resistance
In order to calculate the initial resistance of our silicon wafer, we must first plot the voltage and the current measured across it before finding the slope of that plot. As resistance, R, can be found using Ohm's law:
$$V = IR \to R = \frac{V}{I}$$
For this calculation, we are considering the resistance across the conductor parallel to the direction of the current and without a magnetic field present.

![diagram](images\semiconductor_diagram.png)

Referencing the above diagram of the semiconducting chip, we measured the voltages and currents from $B \to C$ and from $Q \to R$.

```python
import hall_functions
```

```python
Rbc, Rbc_sigma = hall_functions.plot_resistance('Data/BbRcRun1.csv', 'B', 'C', Magnet=False, savefig=True)
```

```python
Rqr, Rqr_sigma = hall_functions.plot_resistance('Data/BqRrRUN2.csv', 'Q', 'R', Magnet=False, savefig=True)
```

```python
avg_R = (Rbc + Rqr)/2
avg_R_sigma = ((Rbc_sigma + Rqr_sigma)/2)
print(f'Average resistance measured across chip = {avg_R:.3f} ohms +/- {avg_R_sigma:.3f} ohms')
```
